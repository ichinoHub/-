```python
import pandas as pd
df = pd.read_excel("売上一覧(改).xlsx",sheet_name ="まとめ")
df = pd.read_excel("売上一覧(改).xlsx",sheet_name ="まとめ",header=2)
df = df.drop(columns = ["Unnamed: 0","Unnamed: 9","Unnamed: 10"])
df = df.dropna()
df
# 担当者ごとの売上金額合計を計算する
sum_by_tantou = df.groupby("担当者")["売上金額"].sum()

# Excelファイルに担当者ごとの売上金額合計を書き込む
tantou_list = df["担当者"].unique()
existing_file ="売上一覧(改).xlsx"
with pd.ExcelWriter(existing_file, mode="a" ,engine="openpyxl") as writer:
      for tantou, sum_value in sum_by_tantou.items():
 # シート名に担当者名と売上金額合計を連結して書き込む
        sheet_name = f"{tantou}_{int(sum_value)}円"
        df_tantou = df[df["担当者"] == tantou]
        df_tantou.to_excel(writer, sheet_name, index=False)
```
